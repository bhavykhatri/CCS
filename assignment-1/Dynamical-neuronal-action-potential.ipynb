```python
def f(x, y, ALPHA):
    if x<=0:
        return ALPHA/(1-x) + y
    elif x>0 and x <ALPHA + y:
        return ALPHA + y
    else:
        return -1
```

```python
ALPHA_LST = [1,2,4,8,10]
MU = 0.001
SIGMA = -1.1
N = 100

x = np.zeros(N)
y =  -3.84
i = 1 
for alpha in ALPHA_LST:
    for i in range(N-1):
        x[i+1] = f(x[i], y, alpha)
        y = y- MU*(x[i] + 1) +MU*SIGMA
    plt.plot(range(N), x)
plt.show()
```
