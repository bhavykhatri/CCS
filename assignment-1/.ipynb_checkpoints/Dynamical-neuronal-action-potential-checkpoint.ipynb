```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y, ALPHA):
    if x<=0:
        return ALPHA/(1-x) + y
    elif x>0 and x <ALPHA + y:
        return ALPHA + y
    else:
        return -1
```

```python
MU = 0.001
N = 2000
alpha_ls = [4, 4, 6]
sigma_ls = [-0.01, 0.01, -0.1]
neural_behavior = ["regime-of-silence", "tonic-spiking", "spiking-bursting"]

x = np.zeros(N)
y =  -3.84
j = 1 
for alpha, sigma,beh in zip(alpha_ls, sigma_ls, neural_behavior):
    plt.figure(j)
    j +=1
    for i in range(N-1):
        x[i+1] = f(x[i], y, alpha)
        y = y- MU*(x[i] + 1) +MU*sigma
    plt.plot(range(N), x)
    plt.xlabel("n - no of iterations")
    plt.ylabel("xn")
    plt.title(beh + " for alpha = " + str(alpha) + " and sigma = " + str(sigma) )
    plt.savefig(beh + ".jpg")
```

```python
alpha_ls = [1, 2, 4, 6, 8]
sigma_ls = [ -1, -0.1,-0.01, 0.01, 0.1, 1]
j=1
for alpha in alpha_ls:
    for sigma in sigma_ls:
        plt.figure(j)
        j +=1
        for i in range(N-1):
            x[i+1] = f(x[i], y, alpha)
            y = y- MU*(x[i] + 1) +MU*sigma
        plt.plot(range(N), x)
        plt.xlabel("n - no of iterations")
        plt.ylabel("xn")
        plt.title("for alpha = " + str(alpha) + " and sigma = " + str(sigma) )
        plt.savefig("DNAP" + str(j) + ".jpg")
```
